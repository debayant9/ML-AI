```python
# Importing libraries
import numpy as np
import random
import math
from collections import deque
import collections
import pickle

# for building DQN model
from keras import layers
from keras import Sequential
from keras.layers import Dense, Activation, Flatten
from keras.optimizers import Adam
```

```python
state_size1 = 36
action_size1 = 21
learning_rate = 0.01
def build_model_arch1():
    model = Sequential()
    # Write your code here: Add layers to your neural nets   
    # hidden layers
    model.add(Dense(32, input_dim=state_size1, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(32, activation='relu', kernel_initializer='he_uniform'))

    # the output layer: output is of size num_actions
    model.add(Dense(action_size1, activation='relu', kernel_initializer='he_uniform'))

    model.compile(loss='mse',optimizer=Adam(lr=learning_rate))
    model.summary
    return model
```

```python
state_size2 = 46
action_size2 = 1
learning_rate = 0.01
def build_model_arch2():
    model = Sequential()
    # Write your code here: Add layers to your neural nets   
    # hidden layers
    model.add(Dense(32, input_dim=state_size2, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(32, activation='relu', kernel_initializer='he_uniform'))

    # the output layer: output is of size num_actions
    model.add(Dense(action_size2, activation='relu', kernel_initializer='he_uniform'))

    model.compile(loss='mse',optimizer=Adam(lr=learning_rate))
    model.summary
    return model
```

```python
#print(state_encod2)
model1 = build_model_arch1()
q = model1.predict(state_encod2)
```

```python
States_track = collections.defaultdict(dict)
print(len(States_track))
```

```python
def initialise_tracking_states():
    sample_q_values = [((1,9,1),(1,2)),((1,11,3),(1,3)),((1,20,5),(2,4)),((4,10,5),(3,1)),((3,13,0),(3,1)),((0,9,6),(1,3))]    #select any 4 Q-values
    for q_values in sample_q_values:
        state = q_values[0]
        action = q_values[1]
        States_track[state][action] = []    #this is an array which will have appended values of that state-action pair for every 2000th episode         
```

```python
def save_tracking_states(qvalues, state):
    print(state)
    if state in States_track.keys():
        action = list(States_track[state].keys())
        print(action)
        ind = ytu.index(action[0])
        print(ind)
        qv = qvalues[0][ind]
        print(qv)
        States_track[state][action[0]].append(qv)
```
